```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from scipy.stats import entropy

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train ID3 (entropy)
model = DecisionTreeClassifier(criterion="entropy", random_state=42)
model.fit(X_train, y_train)

tree = model.tree_

# Root node is always 0
node = 0
feature = tree.feature[node]
threshold = tree.threshold[node]

# Parent entropy
parent_samples = tree.value[node][0]
parent_total = np.sum(parent_samples)
parent_entropy = entropy(parent_samples / parent_total, base=2)

# Left child
left = tree.children_left[node]
left_samples = tree.value[left][0]
left_total = np.sum(left_samples)
left_entropy = entropy(left_samples / left_total, base=2)

# Right child
right = tree.children_right[node]
right_samples = tree.value[right][0]
right_total = np.sum(right_samples)
right_entropy = entropy(right_samples / right_total, base=2)

# Weighted entropy
weighted_entropy = (
    (left_total / parent_total) * left_entropy +
    (right_total / parent_total) * right_entropy
)

# Information Gain
info_gain = parent_entropy - weighted_entropy

print("First Split (Root Node)")
print("Feature:", feature_names[feature])
print("Threshold:", round(threshold, 3))
print("Parent Entropy:", round(parent_entropy, 3))
print("Information Gain:", round(info_gain, 3))
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# CART Regressor
cart = DecisionTreeRegressor(random_state=42)
cart.fit(X_train, y_train)

# Predictions
y_pred = cart.predict(X_test)

# Metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("CART MSE:", mse)
print("CART R2 Score:", r2)
```

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Load dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target
feature_names = housing.feature_names

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train CART
cart = DecisionTreeRegressor(random_state=42)
cart.fit(X_train, y_train)

tree = cart.tree_

# Root node (best split)
node = 0
feature = tree.feature[node]
threshold = tree.threshold[node]

# Parent MSE (impurity in regression tree)
parent_mse = tree.impurity[node]

# Left child
left = tree.children_left[node]
left_mse = tree.impurity[left]
left_samples = tree.n_node_samples[left]

# Right child
right = tree.children_right[node]
right_mse = tree.impurity[right]
right_samples = tree.n_node_samples[right]

# Total samples
total_samples = tree.n_node_samples[node]

# Weighted MSE
weighted_mse = (
    (left_samples / total_samples) * left_mse +
    (right_samples / total_samples) * right_mse
)

print("Best Split (Root Node)")
print("Feature:", feature_names[feature])
print("Threshold:", round(threshold, 3))
print("Parent MSE:", round(parent_mse, 3))
print("Weighted MSE:", round(weighted_mse, 3))
```
